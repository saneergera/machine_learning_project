```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, mean_squared_error, r2_score
from imblearn.over_sampling import SMOTE
import plotly.offline as py
import plotly.graph_objs as go


from sklearn.metrics import roc_curve, auc
from sklearn.multiclass import OneVsRestClassifier
from sklearn.preprocessing import label_binarize
```

```python
fifa_data = pd.read_csv('./pre_processed.csv')
```

```python
fifa_data.drop(columns = ['Unnamed: 0'], inplace=True)
df_pos = fifa_data.copy()
```

```python
X = df_pos.drop(columns=['Position'])
X = pd.get_dummies(X)
y = df_pos['Position']
```

## Training and Testing Models using Train Test Split

```python
#Splitting dataset into train and test sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
def build_classifier (model,name):
    model.fit(X_train, y_train)
    prediction_test = model.predict(X_test)
    prediction_train = model.predict(X_train)
    
    print('Performance Metrics: Test Data')
    print(classification_report(y_test, prediction_test, target_names=['Goalkeepers', 'Defenders', 'Midfielders', 'Strikers']))
    print(confusion_matrix(y_test, prediction_test))
    print('Accuracy Score: ', accuracy_score(y_test, prediction_test))
    print('******************************************************************************************\n')
    print('Performance Metrics: Training Data')
    print(classification_report(y_train, prediction_train, target_names=['Goalkeepers', 'Defenders', 'Midfielders', 'Strikers']))
    print(confusion_matrix(y_train, prediction_train))
    print('Accuracy Score: ', accuracy_score(y_train, prediction_train))
    
    

    return {'Classification type' : name}  
```

## 1) Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dec_tree = DecisionTreeClassifier()
```

```python
build_classifier(dec_tree, 'Decision Tree')
```

## 2) K-Nearest Neighbor

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=5)
```

```python
build_classifier(knn_model, 'K-Nearest Neighbor')
```

## ROC curves for Four classes

```python
def draw_roc(model):
    Y = y

    Y = label_binarize(y, classes=[0,1,2,3])
    classes = 4

    # shuffle and split training and test sets
    X_train_roc, X_test_roc, y_train_roc, y_test_roc = train_test_split(X, Y, test_size=0.33, random_state=0)

    # classifier
    clf = OneVsRestClassifier(model)
    y_score = clf.fit(X_train_roc, y_train_roc).predict_proba(X_test_roc)

    # Compute ROC curve and ROC area for each class
    roc_auc = dict()
    tpr = dict()
    fpr = dict()
    
    for i in range(classes):
        fpr[i], tpr[i], _ = roc_curve(y_test_roc[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

# Plot of a ROC curve for a specific class
    for i in range(classes):
        plt.figure(figsize=(8,6))
        plt.plot(fpr[i], tpr[i], label='ROC curve (area = %0.2f)' % roc_auc[i])
        plt.plot([0, 1], [0, 1], 'k--')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.ylabel('True Positive Rate')
        plt.xlabel('False Positive Rate')
        plt.legend(loc="lower right")
        plt.title('ROC Curve')
        plt.show()
```

```python
draw_roc(gb_clf)
```
